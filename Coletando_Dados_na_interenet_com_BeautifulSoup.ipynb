```python
# Importa o módulo 're' para trabalhar com expressões regulares
import re

#Importa a função 'get' do módulo 'request' para fazer requisições HTTP
from requests import get

# Importa a classe 'BeautifulSoup' do módulo 'bs4' para fazer parsing de HTTP
from bs4 import BeautifulSoup
```

```python
r = get('https://pizzadedados.com') # Faz uma requisição GET para a URL especificada a armazena a resposta na variavle 'r'
```

```python
# Decodifica o conteúdo da resposta para UTF-8 e o transforma em um objeto BeautifulSoup
# Isso permite analisar e manipular o HTML retornado pela requisição
sop = BeautifulSoup(r.content.decode('utf-8'), 'html.parser')

# Imprime os primeiros 400 caracteres do HTML formatado com indentação, para facilitar a leitura.
print(sop.prettify()[0:400])
```

```python
# Encontra todas as tags 'a' (links) no HTML e imprime o valor do atributo 'href' de cada link
for link in sop.find_all('a'):
  print(link.get('href'))
```

```python
# Abre (ou cria) um arquivo de texto chamado 'pizzad_dados.txt' no modo de escrita ('w').
file =  open('pizzad_dados.txt', 'w')

# Define um dicionário com parâmetros de busca que procura por atributos 'href' que começam com 'http'
parametros_busca = {'href': re.compile("^http")}

# Encontra todas as tags 'a' que possuem um atributo 'href' começando com 'http'
# Converte cada link em uma string, imprime o link e o escreve no arquivo 'pizzad_dados.txt', adicionando uma nova linha após cada link
for link in sop.find_all('a', attrs=parametros_busca):
  sopa_link = str(link.get('href'))
  print(sopa_link)
  file.write(sopa_link)
  file.write('\n')

# Garante que todos os dados foram escritos no arquivo e fecha o arquivo
file.flush()
file.close()
```
